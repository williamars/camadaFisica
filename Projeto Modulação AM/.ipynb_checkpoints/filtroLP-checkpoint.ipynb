## Filtro passa-baixa


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.fftpack import fft, fftshift
from suaBibSignal import *
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
import soundfile   as sf
```

```python
def calcFFT(signal, fs):
    # https://docs.scipy.org/doc/scipy/reference/tutorial/fftpack.html
    #y  = np.append(signal, np.zeros(len(signal)*fs))
    N  = len(signal)
    T  = 1/fs
    xf = np.linspace(-1.0/(2.0*T), 1.0/(2.0*T), N)
    yf = fft(signal)
    return(xf, fftshift(yf))
```

```python
def LPF(signal, cutoff_hz, fs):
        from scipy import signal as sg
        #####################
        # Filtro
        #####################
        # https://scipy.github.io/old-wiki/pages/Cookbook/FIRFilter.html
        nyq_rate = fs/2
        width = 5.0/nyq_rate
        ripple_db = 120.0 #dB
        N , beta = sg.kaiserord(ripple_db, width)
        taps = sg.firwin(N, cutoff_hz/nyq_rate, window=('kaiser', beta))
        return( sg.lfilter(taps, 1.0, signal))
```

```python
def generateSin(freq, time, fs):
    n = time*fs #numero de pontos
    x = np.linspace(0.0, time, n)  # eixo do tempo
    s = np.sin(freq*x*2*np.pi)
    #plt.figure()
    #plt.plot(x,s)
    return (x, s)
```

```python
T=2.15
fs=44100
x, y1 = generateSin(100,T,fs)
y1=10*y1
x, y2 = generateSin(4000,T,fs)
x, y3 = generateSin(1000,T,fs)
y=y1+y2 + 4*y3

plt.figure()
plt.xlim(0,T/200)
t=np.linspace(-T/2,T/2,T*fs)
plt.plot(t, y1, 'r')
plt.plot(t, y2, 'b')

plt.figure()
plt.plot(t, y, 'b')
plt.xlim(0,T/20)
```

```python
yFiltrado = LPF(y, 995, fs)
print("filtrado")
#sd.play(yFiltrado)
#sd.wait()
plt.plot(t, yFiltrado, 'b')
plt.xlim(0,0.1)
plt.grid()
plt.title('Filtrado no tempo')
```

```python
#leituara do arquivo audio
fs = 44100   # taxqa de amostagem (sample rate)
sd.default.samplerate = fs
sd.default.channels = 1
audio, samplerate = sf.read('camFis.wav')
yAudio = audio[:,1]
samplesAudio = len(yAudio)
sd.play(audio)
sd.wait()
```

```python
plt.plot(yAudio)
plt.grid()
plt.title('Oi no tempo')
```

```python
X, Y = calcFFT(yAudio, samplerate)
plt.figure("Fourier Audio")
plt.plot(X, np.abs(Y))
plt.grid()
plt.title('Frequencia')
plt.xlim(0,600)
```

```python
yFiltrado = LPF(yAudio, 350, fs)
print("filtrado")
sd.play(yFiltrado)
sd.wait()
plt.plot(yFiltrado)
plt.grid()
plt.title('Filtrado no tempo')
```

```python
X, YF = calcFFT(yFiltrado, samplerate)
plt.figure("Fourier Audio filtrado")
plt.plot(X, np.abs(YF))
plt.grid()
plt.title('Frequencia - filtrado')
plt.xlim(0,1500)
```
